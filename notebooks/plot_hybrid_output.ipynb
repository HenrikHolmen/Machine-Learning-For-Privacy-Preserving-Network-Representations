```python
import sys
import os

code_dir = os.path.abspath(os.path.join(os.getcwd(), "..", "code"))
sys.path.append(code_dir)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

os.makedirs(f"../data/results/plots/hybrid", exist_ok=True)
```

```python
# Load files
n = 100  # change if amount nodes has been changed
K = 5  # change if K-anonymity constrain has been changed
seed = 210659  # change if seed has been changed

filename_suffix = f"n{n}_K{K}_seed{seed}"
block_sizes = np.load(f"../data/results/hybrid/final_block_sizes_{n}_K{K}.npy")
```

```python
# Plot block size histogram
plt.figure(figsize=(6, 4))
sns.histplot(
    block_sizes,
    bins=range(min(block_sizes), max(block_sizes) + 2),
    discrete=True,
    shrink=0.8,
    color="skyblue",
    edgecolor="black",
)

plt.title(f"Final Block Sizes Histogram (K={K}) - {len(block_sizes)} Blocks")
plt.xlabel("Block Size")
plt.ylabel("Number of Blocks")
plt.xticks(range(min(block_sizes), max(block_sizes) + 1))
plt.grid(True, axis="y", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.savefig(
    f"../data/results/plots/hybrid/final_block_sizes_n{n}_K{K}_seed{seed}.png",
    dpi=500,
    bbox_inches="tight",
)
plt.show()
```
